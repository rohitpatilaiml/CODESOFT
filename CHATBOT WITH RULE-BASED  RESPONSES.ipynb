RESPONSES

```python
import random

# Define predefined responses for different user inputs
responses = {
    "hi": ["Hello!", "Hi there!", "Hey!"],
    "how are you": ["I'm good, thank you!", "I'm doing well, thanks for asking!"],
    "bye": ["Goodbye!", "See you later!", "Bye!"],
    "thank you": ["You're welcome!", "No problem!", "Glad I could help!"],
    "help": ["I'm here to assist you with any questions or concerns. Feel free to ask!", "Sure, what do you need help with?"],
    "what's your name": ["I'm a chatbot designed to assist you!", "I'm just a humble chatbot!"],
    "who created you": ["I was created by a team of developers!", "My creators prefer to remain anonymous!"],
    "tell me a joke": ["Why don't scientists trust atoms? Because they make up everything!", "What did one ocean say to the other ocean? Nothing, they just waved!"],
    "how old are you": ["I'm ageless!", "I don't have an age, but I'm here to help you!"],
    "what is the meaning of life": ["The meaning of life is subjective and varies from person to person!", "That's a deep question!"],
    "what can you do": ["I can answer your questions, provide assistance, or just chat with you!", "I'm here to help you with anything you need!"],
    "tell me a fun fact": ["Did you know that the shortest war in history lasted only 38 minutes?", "Bananas are berries, but strawberries aren't!"],
    "tell me about yourself": ["I'm a chatbot designed to assist users like you!", "I'm programmed to answer your questions and provide assistance!"],
    "how do I contact support": ["You can contact support via email or phone. Let me know if you need assistance with that!", "I can provide you with contact information for support!"],
    "can you recommend a movie": ["Sure! How about watching 'The Shawshank Redemption'?", "Have you seen 'Inception'? It's a great movie!"],
    "what's the weather like today": ["I'm not sure, but you can check the weather forecast online or on your phone!", "I can't check the weather, but you can easily find that information online!"],
    "tell me about artificial intelligence": ["Artificial intelligence (AI) is the simulation of human intelligence in machines!", "AI is a branch of computer science that aims to create intelligent machines!"],
    "can you play a game": ["I'm not equipped to play games, but I'm here to chat with you!", "Unfortunately, I'm not a game-playing bot, but I'm happy to chat with you!"],
    "what's your favorite color": ["I don't have a favorite color, but I do like chatting with users like you!", "I don't have preferences like humans do, but I'm here to assist you!"],
    "tell me a story": ["Once upon a time, in a faraway land, there was a chatbot who loved to chat with users!", "Here's a story: Once there was a chatbot who wanted to learn everything about the world!"],
    "tell me about your hobbies": ["As a chatbot, I don't have hobbies, but I enjoy assisting users like you!", "I'm a chatbot designed to help users, so chatting with you is my hobby!"],
    "what's the time": ["I don't have access to real-time information like the current time, but you can check your device's clock!", "Unfortunately, I can't provide real-time updates like the current time, but I'm here to assist you with other questions!"],
    "default": ["I'm sorry, I didn't understand that.", "Could you please rephrase that?", "I'm not sure what you mean."]
}

# Function to get response based on user input
def get_response(user_input):
    # Convert user input to lowercase for case-insensitive matching
    user_input = user_input.lower()

    # Check if user input matches any predefined patterns
    for key in responses:
        if key in user_input:
            return random.choice(responses[key])

    return random.choice(responses["default"])

# Main function to interact with the chatbot
def main():
    print("Welcome to the Simple Chatbot!")
    print("You can start chatting. Enter 'bye' to exit.")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'bye':
            print(get_response(user_input))
            break
        else:
            print("Bot:", get_response(user_input))

# Run the chatbot
if __name__ == "__main__":
    main()
```
